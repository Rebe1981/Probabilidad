```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#datos de entrada
publicidad = np.array([2, 3, 5, 4, 11, 5]).reshape(-1, 1)
ganancias = np.array([20, 25, 34, 30, 40, 31])
```

```python
#crear y entrenar el modelo de regresion lineal
modelo = LinearRegression()
modelo.fit(publicidad, ganancias)
```

```python
#predecir la ganancia para la inversion de 8 en publicidad
inversion_prediccion = np.array([[8]])
ganancia_predicha = modelo.predict(inversion_prediccion)
```

```python
print("La ganancia esperada para una inversionn de 8 en publicidad es de: ",ganancia_predicha[0])
print("Valor de la intercepcion: ", modelo.intercept_)
print("Valor de la pendiente: ", modelo.coef_)
```

```python
#crear el diagrama de dispersion con la linea de regresion
plt.figure(figsize=(6, 5))
plt.scatter(publicidad, ganancias, color='blue', label='Datos')
plt.plot(publicidad, modelo.predict(publicidad), color='red', label='Linea de regresion')
plt.scatter(inversion_prediccion, ganancia_predicha, color='green', marker='x', s=100, label='Prediccion (8, {0:.2f})'.format(ganancia_predicha[0]))

plt.title('Pronostico de ganancias a partir de la inversion en publicidad')
plt.xlabel('Inversion en publicidad(millones)')
plt.ylabel('Ganancias anuales(millones)')
plt.legend()
plt.grid(True)
plt.show()
```
