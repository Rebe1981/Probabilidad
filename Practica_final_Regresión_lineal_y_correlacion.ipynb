```python
import matplotlib.pyplot as plt
from statistics import*
import pandas as pd
import math
from google.colab import drive
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
```

```python
grupo1=pd.read_csv('grupo1.csv')
grupo2=pd.read_csv('grupo2.csv')
grupo3=pd.read_csv('grupo3.csv')
grupo1.describe()
grupo2.describe()
grupo3.describe()
```

```python
plt.hist(grupo1['Calificacion'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Calificaciones')
plt.ylabel('Frecuencia')
plt.title('Histograma de Calificaciones del grupo 1')
```

```python
plt.hist(grupo2['Calificacion'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Calificaciones')
plt.ylabel('Frecuencia')
plt.title('Histograma de Calificaciones del grupo 2')

```

```python
plt.hist(grupo3['Calificacion'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Calificaciones')
plt.ylabel('Frecuencia')
plt.title('Histograma de Calificaciones del grupo 3')
```

```python
x = grupo1[['HorasEstudio']]
y = grupo1['Calificacion']
model = LinearRegression()
model.fit(x, y)
pendiente = model.coef_[0]
intercepto = model.intercept_
print("Ecuación de la línea de Regresión:")
print(f"Calificaciones = {pendiente:.2f} * Horas de Estudio + {intercepto:.2f}")
plt.scatter(grupo1['HorasEstudio'], grupo1['Calificacion'], color='blue', label='Datos')
plt.plot(grupo1[['HorasEstudio']], model.predict(x), color='red', label='Línea de Regresión')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Regresión Lineal= Horas de Estudio VS Calificaciones del grupo 1')
plt.legend()
plt.show()

coef_corr, _ = pearsonr(grupo1['HorasEstudio'], grupo1['Calificacion'])
print("Coeficiente de correlación:",coef_corr)
```

```python
x = grupo2[['HorasEstudio']]
y = grupo2['Calificacion']
model = LinearRegression()
model.fit(x, y)
pendiente = model.coef_[0]
intercepto = model.intercept_
print("Ecuación de la línea de Regresión:")
print(f"Calificaciones = {pendiente:.2f} * Horas de Estudio + {intercepto:.2f}")
plt.scatter(grupo2['HorasEstudio'], grupo2['Calificacion'], color='blue', label='Datos')
plt.plot(grupo2[['HorasEstudio']], model.predict(x), color='red', label='Línea de Regresión')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Regresión Lineal: Horas de Estudio VS Calificaciones del grupo 2')
plt.legend()
plt.show()

coef_corr, _ = pearsonr(grupo2['HorasEstudio'], grupo2['Calificacion'])
print("Coeficiente de correlación: ",coef_corr)
```

```python
x = grupo3[['HorasEstudio']]
y = grupo3['Calificacion']
model = LinearRegression()
model.fit(x, y)
pendiente = model.coef_[0]
intercepto = model.intercept_
print("Ecuación de la línea de Regresión:")
print(f"Calificaciones = {pendiente:.2f} * Horas de Estudio + {intercepto:.2f}")
plt.scatter(grupo3['HorasEstudio'], grupo3['Calificacion'], color='blue', label='Datos')
plt.plot(grupo3[['HorasEstudio']], model.predict(x), color='red', label='Línea de Regresión')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Regresión Lineal: Horas de Estudio VS Calificaciones')
plt.legend()
plt.show()

coef_corr, _ = pearsonr(grupo3['HorasEstudio'], grupo3['Calificacion'])
print("Coeficiente de correlación:",coef_corr)
```
