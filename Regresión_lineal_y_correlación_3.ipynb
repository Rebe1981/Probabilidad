```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#definir tama√±o
plt.figure(figsize=(6, 4))
```

```python
#datos de entrada
inversion = [2, 3, 5, 4, 11, 5]
ganancias = [20, 25, 34, 30, 40, 31]
```

```python
#crear el diagrama de dispersion con linea de regresion
sns.set(style="whitegrid")
sns.regplot(x=inversion, y=ganancias, scatter=True, color='blue')
plt.title('Diagrama de Dispersion con Linea de Regresion')
plt.xlabel('Inversion(millones)')
plt.ylabel('Ganancias(millones)')
plt.grid(True)
plt.show()
```
