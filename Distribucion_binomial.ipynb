```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
media=70
desv=4
Total_a=500
lim=79
x1=68
x2=72
prob_72=norm.cdf(lim,media,desv)
prob_68=norm.cdf(lim,media,desv)
probest_mas68=1-prob_68
probest_menos72=prob_72*Total_a
print(probest_mas68)
print(probest_menos72)
x=np.linspace(media-3*desv,media+3*desv,Total_a)
pdf=1/(desv*np.sqrt(2*np.pi))*np.exp(-(x-media)**2/(2*desv**2))
x=np.linspace(media-4*desv,media+4*desv,Total_a,1000)
y=norm.pdf(x,media,desv)
```

```python
x=np.linspace(media-4*desv,media+4*desv,Total_a)
plt.plot(x,norm.pdf(x,media,desv))
y=norm.pdf(x,media,desv)
```

```python
plt.plot(x,y,'b',linewidth=2)
plt.fill_between(x,y,where=(x>=x1),color='skyblue',alpha=0.5)
plt.title('Probabilidad de que un estudiante pese mas de 68kg.\n$\mu=70$,$\sigma=4$,x=500')
plt.xlabel("Peso en kg")
plt.ylabel("Densidad de probabilidad")
plt.grid(True)
```

```python
plt.plot(x,y,'b',linewidth=2)
plt.fill_between(x,y,where=(x<=x2),color='red',alpha=0.5)
plt.title('Probabilidad de que un estudiante pese mas de 72kg.\n$\mu=70$,$\sigma=4$,x=500')
plt.xlabel("Peso en kg")
plt.ylabel("Densidad de probabilidad")
plt.grid(True)
```

```python
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=((x >= x1) & (x <= x2)), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese entre 68kg y 72kg.\n$\mu=70$, $\sigma=4$, n=500')
plt.xlabel('Peso en kg')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
