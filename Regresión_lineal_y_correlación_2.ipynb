```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#definir tama√±o de la figura
plt.figure(figsize=(6, 4))
```

```python
#datos de entrada
inversion = [2, 3, 5, 4, 11, 5]
ganancias = [20, 25, 34, 30, 40, 31]
```

```python
#dibujar el diagrama de dispersion
sns.scatterplot(x = inversion, y = ganancias)
plt.title('Diagrama de Dispersion de Inversion VS Ganancias')
plt.xlabel('Inversion(millones)')
plt.ylabel('Ganancias(millones)')
plt.grid(True)
plt.show()
```
